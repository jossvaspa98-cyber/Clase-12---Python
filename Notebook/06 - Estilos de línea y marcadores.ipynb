# Estilos de línea y marcadores
Objetivos:
- Variar estilo de línea, color y marcador.
- Construir una pequeña paleta consistente.
- Plantear un reto breve al final.


```python
# Imports y estilo
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("ggplot")
plt.rcParams["figure.figsize"] = (9, 4)
np.random.seed(7)
```

## Comparativa de estilos
Usamos diferentes combinaciones para distinguir series sin depender solo del color.


```python
puntos = np.arange(0, 10)
base = np.linspace(10, 40, len(puntos))

estilos = [
    ("tab:blue", "-", "o", "Serie A"),
    ("tab:orange", "--", "s", "Serie B"),
    ("tab:green", ":", "^", "Serie C"),
]

fig, ax = plt.subplots()
for color, linea, marcador, nombre in estilos:
    ruido = np.random.randint(-5, 5, size=len(puntos))
    ax.plot(puntos, base + ruido, color=color, linestyle=linea, marker=marcador, label=nombre)

ax.set_title("Estilos de línea y marcador")
ax.set_xlabel("Índice")
ax.set_ylabel("Valor")
ax.legend()
fig.tight_layout()
```

## Ejercicio guiado
- Cambia el grosor (`linewidth`) y la transparencia (`alpha`).
- Usa marcadores solo cada 2 puntos con `markevery=2`.


```python

fig, ax = plt.subplots()
ax.plot(puntos, base, color="tab:purple", linestyle="-", marker="o", linewidth=3, alpha=0.7, label="Línea gruesa")
ax.plot(puntos, base - 5, color="tab:cyan", linestyle="--", marker="D", markevery=2, label="Marcadores espaciados")
ax.set_title("Probando grosor y transparencia")
ax.legend()
fig.tight_layout()
```

## Resumen
- Combinar color + estilo + marcador evita ambigüedades.
- `linewidth`, `alpha` y `markevery` afinan la legibilidad.
